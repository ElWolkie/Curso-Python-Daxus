## Generando archivos PDF del presupuesto

* Definir un tipo de fuente
* Llenamos el documento
* Guardar el documento

```python
from fpdf import FPDF
```

```python
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

pdf.image("Template.png", 0, 0)

pdf.text(110, 150, proyecto)
pdf.text(110, 160, horas_estimadas)
pdf.text(110, 170, valor_hora)
pdf.text(110, 180, plazo_estimado)
pdf.text(110, 190, f"El valor total del proyecto es: {resultado}")

pdf.output("presupuesto.pdf")
```
